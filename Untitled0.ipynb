```python
import numpy as np
x = np.array([1,7,2,3,2,2,2,4,4,7,3,2,1])
print("Original array:")
print(x)
  
print("Most frequent value in the above array:")
print(np.bincount(x).argmax())
```

```python
import numpy as np
  
  
x = np.array([1,1,1,1,1,2,3,3,3,3,3])
print("Original array:")
print(x)
  
print("Most frequent value in above array")
y = np.bincount(x)
print(y)
maximum = max(y)
  
for i in range(len(y)):
    if y[i] == maximum:
        print(i, end=" ")
```
